```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
file_path = 'C:/Users/ASUS/OneDrive/Рабочий стол/Data_to_Transform.csv'
df = pd.read_csv('C:/Users/ASUS/OneDrive/Рабочий стол/Data_to_Transform.csv')

df
```

```python
def plot_qq_plots(data, title):
    plt.figure(figsize=(10, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.show()
```

```python
plot_qq_plots(df['Moderate Positive Skew'], "Original Data - Moderate Positive Skew")
plot_qq_plots(df['Highly Positive Skew'], "Original Data - Highly Positive Skew")
plot_qq_plots(df['Moderate Negative Skew'], "Original Data - Moderate Negative Skew")
plot_qq_plots(df['Highly Negative Skew'], "Original Data - Highly Negative Skew")
```

```python
df_log = np.log(df)
plot_qq_plots(df_log['Moderate Positive Skew'], "Log-transformed Data - Moderate Positive Skew")
plot_qq_plots(df_log['Highly Positive Skew'], "Log-transformed Data - Highly Positive Skew")
plot_qq_plots(df_log['Moderate Negative Skew'], "Log-transformed Data - Moderate Negative Skew")
plot_qq_plots(df_log['Highly Negative Skew'], "Log-transformed Data - Highly Negative Skew")
```

```python
df_cbrt = np.cbrt(df)
plot_qq_plots(df_cbrt['Moderate Positive Skew'], "Cubic Root-transformed Data - Moderate Positive Skew")
plot_qq_plots(df_cbrt['Highly Positive Skew'], "Cubic Root-transformed Data - Highly Positive Skew")
plot_qq_plots(df_cbrt['Moderate Negative Skew'], "Cubic Root-transformed Data - Moderate Negative Skew")
plot_qq_plots(df_cbrt['Highly Negative Skew'], "Cubic Root-transformed Data - Highly Negative Skew")
```

```python
df_sqrt = np.sqrt(df)
plot_qq_plots(df_sqrt['Moderate Positive Skew'], "Square Root-transformed Data - Moderate Positive Skew")
plot_qq_plots(df_sqrt['Highly Positive Skew'], "Square Root-transformed Data - Highly Positive Skew")
plot_qq_plots(df_sqrt['Moderate Negative Skew'], "Square Root-transformed Data - Moderate Negative Skew")
plot_qq_plots(df_sqrt['Highly Negative Skew'], "Square Root-transformed Data - Highly Negative Skew")
```
